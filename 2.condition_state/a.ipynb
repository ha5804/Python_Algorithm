# __조건문 활용__

## `NO.2525`

### __Function__
- (1)input() 함수는 사용자로부터 `문자열`을 입력받는다.
- (2)split() 함수는 공백 기준으로 문자열을 나눈다. `결과는 리스트`
- (3)map()함수는 여러 개의 데이터를 한번에 다른 형태로 변환한다.

### __Algorithm__
- (1)시간은 전체 걸린 분의 60에 대한 몫을 더한다.
- (2)분은 전체 걸린 분의 60에 대한 나머지 이다.

### __error__
- (1)m 수정후 더하는 오류 발생
- (2)total에서 계산하는 방식으로 수정.



````python
h, m = map(int, input().split())
need_m = int(input())

total = m + need_m
h += total // 60
m = total % 60

if (h>=24):
    h -= 24

print(h, m)
````

## `NO.2480`

### __function__
- (1)max() 가장 큰 숫자 반환
- (2)if 조건문 설명
    - 첫 조건에서 (a == b == c)로 설정하여, 그 다음 조건문에서는, 세개다 같을 경우를 예외처리 할 필요없다. 

    ``` python
    if (a == b) and (b != c) 
    #어차피 같으면 첫 조건에서 걸리기 때문.
    ```


````python
a, b, c = map(int, input().split())

if (a == b == c):
    print(10000 + (a * 1000))
elif (a == b) or (a == c):
    print(1000 + (a * 100))
elif (b == c):
    print(1000 + (b * 100))
else:
    print(max(a,b,c) * 100)
````
