# __Iteration station__

## __for문 빠른 입출력__

- for문을 사용할때, 더 빠른 시간을 위해 input()대신에 `sys.stdin.readline()` 함수를 사용하자.



```python
import sys

a = int(sys.stdin.readline().rstrip())
for i in range(a):
    b , c = map(int, sys.stdin.readline().rstrip().split())
    #rstrip으로 문자열 정리후, 리스트로 변환하는 순서여야한다.
    print(b + c)
```


## __별 찍기__
- Algorithm
    - (1)출력 모양을 확인하자.
    - (2)별이 하나씩 늘고, 공백이 하나씩 줄고 있음을 알 수 있다.
    - (3)주어진 숫자 = 별 + 공백
    - (4)별 = 주어진 숫자 - 공백
    - (5)공백 = 주어진 숫자 - 별

```python
a = int(input())
for i in range(1, a+1):
    for j in range(a - i):
        print(" ", end ="")
    print("*" * i)
```

## __입력 종료될때까지 연산하기__

- try, except구문울 사용하여 해결한다.

```python
while 1: #1은 True이다. 즉 True(에러가 안날때까지)
    try:
        A, B = map(int, input().split())
        print(A+B)
    except:
        break
```
