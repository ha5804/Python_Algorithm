### 정수의 자리수 연산은, string으로 변환후 실행한다.

`input()은 기본적으로 string 처리 함수`

```python
a = int(input())
b = input()

print(a * int(b[2]))
print(a * int(b[1]))
print(a * int(b[0]))
print(a * int(b))
```

### 공백을 기준으로 입력 받을때는 split()함수 사용.

```python
a, b = input().split()

print(a, b)
```
